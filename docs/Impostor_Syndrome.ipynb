## What is impostor syndrome?

Impostor Syndrome is the feeling that:
- You are not as capable as others think
- Your success is due to luck
- You might be “exposed” as not good enough

It happens to:
- Beginners
- Intermediate learners
- Even experienced engineers

You are not alone.

## Common thoughts

You might have said:

- “Everyone understands this except me”
- “I’m too slow”
- “I can’t build anything on my own”
- “Maybe tech is not for me”

Pause for a moment.

Have you felt any of these?

```python
feeling = input("Be honest: What doubt or negative thought have you had about yourself? ")

print("\nYou wrote:")
print(feeling)
```

## Let’s reframe that thought

Instead of saying:
“I’m not good at this”

Say:
“I’m still learning this”

Instead of:
“I don’t understand anything”

Say:
“I don’t understand this yet”

Small shift. Big impact.

```python
reframe = input("Rewrite your doubt into a more positive or growth-focused statement: ")

print("\nNew mindset:")
print(reframe)
```
