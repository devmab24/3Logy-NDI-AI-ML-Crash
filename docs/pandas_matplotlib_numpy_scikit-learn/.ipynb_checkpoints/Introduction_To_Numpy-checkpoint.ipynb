The term `broadcasting` describes how NumPy treats arrays with different shapes during arithmetic operations.

Looking at the shape of the data above, the issue is a structural mismatch.
- `a2` has 2 rows and 3 columns.
- `a3` has 2 blocks, each containing 3 rows and 3 columns.
When you try to combine them, NumPy looks at the middle dimension and sees 2 (from a2) vs 3 (from a3) and gives up.
### The Fix:
To make them "talk" to each other, you usually want to treat each row of a2 as a set of multipliers (or add-ons) for each corresponding block in a3. To do that, you need to turn a2 into a (2, 1, 3) shape so the "1" can stretch to match the "3" in a3.

```python
# Use np.newaxis to turn (2, 3) into (2, 1, 3)
result = a2[:, np.newaxis, :] * a3
result
```

## Aggregation
Performing the same operation on a number of things e.g numpy arrays.

```python
#Lets create a list of list using `python`
my_list = [2, 8, 2, 9, 10]
type(my_list)
```

```python
#adding with python `.sum()` function
sum(my_list)
```

```python
#Lets take `a1` as example
type(a1)
```

```python
print(sum(a1))
```

```python
print(np.sum(a1))
```

What do you think is the main difference between these two functions above?
- Use Python's method (`.sum()`) on Python datatypes, and Use Numpy's method on Numpy arrays (`np.sum()`).

```python
#Lets see an example; Create a large random data
large_data = np.random.rand(1000)
large_data.size
```

```python
#Viewing the first 10 elements
large_data[:10]
```

```python
%timeit sum(large_data) #Python's sum `sum()`
%timeit np.sum(large_data) #Numpy's sum (`.np.sum()`)
```

The first line took 64.9 microseconds per loop while the second line too just 3.14 microseconds. So, `64.9/3.14` shows that `Numpy's sum` is 20.6687898089172 faster than `Python's sum `sum()

**Note**: Numpy is been optimized to perform numerical calculations. Whenever you want to perform a numerical calculations make sure to use `numpy's version of an aggregation functions.`

```python
a2
```

```python
#Mean
print(np.mean(a2))
```

```python
#Min
print(np.min(a2))
```

```python
#max
print(np.max(a2))
```

```python
#Variance?: Measure of the average degree to which each number is different to the mean.
#Higher Variance = wider range of numbers
#Lower Variance = lower range of numbers
print(np.var(a2))
```

```python
#STD?: Is the measure of how spread out a group of numbers is from the mean, Its just a square root of the variance
#Standard Deviation = Square root of the variance
print(np.std(a2))
```

```python
#OR
print(np.sqrt(np.var(a2)))
```

```python
#Demo of var & std
high_var = np.array([1, 100, 200, 300, 1000, 4000, 5000])
low_var = np.array([2, 4, 6, 8, 10])
```

```python
print(np.var(high_var), np.var(low_var))
```

```python
print(np.std(high_var), np.std(low_var))
```

```python
print(np.mean(high_var), np.mean(low_var))
```
