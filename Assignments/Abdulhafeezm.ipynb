
# Assignment: Setting Up the Development Environment

This guide outlines the standard procedure for initializing an isolated Python environment using **Anaconda** and launching a **Jupyter Notebook** server for data science and machine learning projects.

---

## 1. Opening the Anaconda Prompt

The **Anaconda Prompt** is a command-line interface (CLI) specifically configured with the Conda package manager. Unlike the standard Windows Command Prompt, it automatically initializes the `base` environment variables required to run Python.

### **Procedure**

1. Click the **Start Menu** in Windows.
2. Search for **"Anaconda Prompt"**.
3. Click the application to open the terminal.

> **Note:** Upon opening, you will see `(base)` at the beginning of the command line, indicating you are in the default root environment.

---

## 2. Navigating to the Project Directory

Before launching any tools, it is best practice to move the terminal focus to your specific project folder. This ensures all notebooks created are saved in the correct location.

### **The `cd` Command**

The `cd` (Change Directory) command is used to navigate the file system. Use the following command to move to your desktop project folder:

```bash
cd C:\Users\strongs_\Desktop\AI_ML001

```

---

## 3. Activating the Conda Environment

To prevent dependency conflicts between different projects, we use isolated environments. In this case, we are activating an environment located within the project folder.

### **Activation Command**

Execute the following to activate your local environment:

```bash
conda activate C:\Users\strongs_\Desktop\AI_ML001\env

```

**Why we do this:**

* **Isolation:** Ensures that libraries installed for Project A don't break Project B.
* **Consistency:** Guarantees you are using the exact version of Python (e.g., 3.10 or 3.11) required for your code.

---

## 4. Launching Jupyter Notebook

Once the environment is active, you can start the web-based development interface.

### **Command**

```bash
jupyter notebook

```

### **System Actions**

When this command is executed, the following happens:

* A **local web server** is initiated on your computer.
* The server defaults to port **8888**.
* Your default web browser will open automatically to: `http://localhost:8888`

---

## 5. Creating a New Notebook

Inside the Jupyter web interface, follow these steps to start coding:

1. Navigate to the top right corner and click **New**.
2. Select **Python 3 (ipykernel)** from the dropdown menu.
3. A new browser tab will open with an empty `.ipynb` file ready for input.

---

## 6. Proper Shutdown Procedure

Closing the browser tab **does not** stop the server. To shut down the environment correctly:

1. Save and close your notebook in the browser.
2. Return to the **Anaconda Prompt** window.
3. Press **`CTRL + C`** on your keyboard.
4. When prompted `Terminate batch job (Y/N)?`, type **`Y`** and press **Enter**.
