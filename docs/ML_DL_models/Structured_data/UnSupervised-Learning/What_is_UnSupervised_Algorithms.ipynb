## UnSupervised Algorithms

### What is an Unsupervised machine learning?
`Unsupervised machine learning` is the process of extracting **patterns** from data. It deals with unlabeled data sets. This form of ML allows algorithms to work on their own to discover information from data, without any guidance. The primary tasks of the Algols here is to classify unsorted informagtion according to patterns, differences, or similarities. `Example` a categorization could be done based on `age, average time spent in hospital, types of disease, etc`.
The final categorization of data set helps to derive some conclusions based on different patterns generated from the data.

### Why Unsupervised machine learning?
Learning from the data is the goal of `Unsupervised machine learning`. The machine derives insight from the data without being thought anything.
- Extracts many kinds of previously unknown patterns in a data set.
- Helps find features that may be useful for categorization.

### Types of Unsupervised Algorithms?
Unsupervised Algorithms extracts patterns from unlabeled or uncategorized data. They used clustering and association techniques help to find the structure of a dataset and group them based on similarity.

### CLUSTERING?
Clustering aims to divides a set of data points into multiple clusters such that the data point withing or cluster are similar to each other. In essence, clustering aims to segregate data points with similar traits.

### ASSOCIATION?
An association rule involves finding the relationships between variables in a large data set. It determines the dependencies of one data item over another data item. Association rule learning often finds it's application in market basket analysis.

### Commonly Used Algorithms?
**Unsupervised Learning Algorithms are:**
- K-Means clustering
- K-memoids
- Desity-based spatial clustering of application with noise (DBSCAN)
- Principal Component Analysis (PCA)
- Agglomerative clustering
- BIRCH

## K-Means clustering
* This is an iterative clustering algorithm that groups the dataset into `k` pre-defined non-pverlapping clusters.
---
## K-medoids
* Is similar to k-means but with the requirments that `cluster centers coincide with points in the data`
---
## DBSCAN
* This algorithms finds core samples of high density and expands clusters from them. It is suitable for a dataset containing clusters of similar density.
---
## PCA
* This algorithms identifies patterns in a dataset and simplifies the dataset by reducing dimensionality.
---
## Agglomerative clustering
* This algorithms forms a hierarchy of clusters by treating each data point as a seperate cluster, this is repeated iteratively untill all the clusters have been merged into one big cluster.
---
## BIRCH
* Is a memory efficient online-learning algorithm. It constructs a tree data structure with cluster centroids being read off the leaf.

### When to use Unsupervised Algorithms?
`Unsupervised Algorithms` are useful when the goal is to extract data from unlabelled data and the output is `NOT` known. They are mainly used when:
* Data points must be grouped based on similar traits.
* The relationship between different features of large dataset should be determined
* The Data input in a large dataset with many features must be reduced to a manageable size i.e the dimensionality of the input data must be reduced.
* Anomalies in dataset should be detected.
---
### Applications of Unsupervised Learning Algorithms?
Unsupervised Learning Algorithms has many real-world applications including:
* *Market basket analysis*
* *Fraud detection*
* *Malware detection*
* *Customer Segmentation*
* *Targeted marketting campaigns*

Thus, the ability of `Unsupervised Algorithms` to group unsorted data based on patterns, similarities and differences make them an ideal choice for application.
