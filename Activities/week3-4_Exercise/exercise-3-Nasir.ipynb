```python
# Import dataset with missing values
car_sales_missing = pd.read_csv("car_sales_csv_format.csv")
```

```python
# Fill missing 'quantity_sold' with mean
car_sales_missing["quantity_sold"].fillna(car_sales_missing["quantity_sold"].mean(), inplace=True)
```

```python
# Remove other missing data
car_sales_missing.dropna(inplace=True)
print("DataFrame after handling missing values:")
print(car_sales_missing.head())
print("\n")
```

```python
# Seats
car_sales_missing["Seats"] = 5
```

```python
# Engine size (random 1.3 to 4.5)
car_sales_missing["Engine_Size"] = np.random.uniform(1.3, 4.5, size=len(car_sales_missing))
```

```python
# 1. Remove any dollar signs, commas, or spaces
car_sales_missing["price_usd"] = car_sales_missing["price_usd"].astype(str)  # ensure string first
car_sales_missing["price_usd"] = car_sales_missing["price_usd"].str.replace(r"[$, ]", "", regex=True)
```

```python
# 2. Convert to numeric safely (handles errors like empty strings)
car_sales_missing["price_usd"] = pd.to_numeric(car_sales_missing["price_usd"], errors="coerce")
```

```python
# Safe way to fill NaN and avoid chained assignment warning
car_sales_missing["price_usd"] = car_sales_missing["price_usd"].fillna(0)
```

```python
# 4. Convert to int
car_sales_missing["price_usd"] = car_sales_missing["price_usd"].astype(int)
```

```python
# Price per km (example calculation)
car_sales_missing["price_per_km"] = car_sales_missing["price_usd"] / 100000
print("DataFrame with new columns:")
print(car_sales_missing.head())
print("\n")
```

```python
# Remove the last column
car_sales_missing.drop("price_per_km", axis=1, inplace=True)
```

```python
# 20. Shuffle the DataFrame
shuffled = car_sales_missing.sample(frac=1).reset_index(drop=True)
print("Shuffled DataFrame:")
print(shuffled.head())
```
