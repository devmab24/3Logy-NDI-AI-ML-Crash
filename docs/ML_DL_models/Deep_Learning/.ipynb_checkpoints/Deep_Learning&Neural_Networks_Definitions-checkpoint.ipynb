## 1. Introduction to Deep Learning

### What is Deep Learning?

Deep Learning is a subset of Machine Learning that uses **Artificial Neural Networks (ANNs)** with multiple layers to model complex patterns in data.

It is especially powerful for:

* Image recognition
* Natural Language Processing (NLP)
* Speech recognition
* Time-series forecasting

### Why Deep Learning?

Traditional ML struggles with:

* High-dimensional data
* Unstructured data (images, text, audio)

Deep Learning excels because it:

* Automatically extracts features
* Scales with more data
* Learns hierarchical representations

## 2. Neural Networks (ANNs)

### Basic Structure

A neural network consists of:

* **Input Layer**
* **Hidden Layers**
* **Output Layer**

Each neuron performs:
[
z = wx + b
]
[
a = f(z)
]

### Key Components

* Weights (w)
* Bias (b)
* Activation functions

### Common Activation Functions

* ReLU
* Sigmoid
* Tanh
* Softmax


### Simple Example (Keras)

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

## 3. Training a Neural Network

### Steps

1. Forward Pass
2. Loss Computation
3. Backpropagation
4. Weight Update (Optimizer)

### Loss Functions

* Regression → MSE
* Classification → CrossEntropy

### Optimizers

* SGD
* Adam (recommended for beginners)

## 4. Convolutional Neural Networks (CNNs)

### What are CNNs?

CNNs are specialized for **image data**.

### Key Concepts

* Convolution
* Filters/Kernels
* Pooling
* Feature maps

### Architecture

* Conv Layer → ReLU → Pooling → Fully Connected

### Example (TensorFlow/Keras)

```python
model = keras.Sequential([
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPooling2D(),
    keras.layers.Flatten(),
    keras.layers.Dense(10, activation='softmax')
])
```

### Applications

* Image classification
* Object detection
* Medical imaging

## 5. Recurrent Neural Networks (RNNs)

### What are RNNs?

RNNs are designed for **sequential data**.

### Key Idea

They have memory of previous inputs.

### Variants

* Vanilla RNN
* LSTM (Long Short-Term Memory)
* GRU (Gated Recurrent Unit)

### Example (Keras)

```python
model = keras.Sequential([
    keras.layers.LSTM(64),
    keras.layers.Dense(1)
])
```

### Applications

* Text generation
* Time series forecasting
* Chatbots

## 6. PyTorch vs TensorFlow

| Feature     | TensorFlow/Keras | PyTorch  |
| ----------- | ---------------- | -------- |
| Ease of Use | High             | Medium   |
| Flexibility | Medium           | High     |
| Debugging   | Harder           | Easier   |
| Popular Use | Production       | Research |

### PyTorch Example

```python
import torch
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(10, 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)
```

## 7. Model Evaluation

### Metrics

* Accuracy
* Precision
* Recall
* F1 Score
* RMSE / MAE

## 8. Overfitting & Regularization

### Overfitting

Model memorizes instead of learning.

### Solutions

* Dropout
* Early stopping
* More data
* Data augmentation

## 9. Best Practices

* Normalize your data
* Start simple
* Use validation set
* Monitor training curves
* Use pretrained models when possible

## 10. Design Strategy (VERY IMPORTANT)

### Beginner Workflow

1. Understand the problem
2. Start with baseline model
3. Add complexity gradually
4. Evaluate
5. Tune hyperparameters

### Key Advice

* Don't jump into CNNs/RNNs without mastering basics
* Debug small models first
* Always visualize results

## 11. Advantages of Deep Learning

* High accuracy
* Handles complex data
* Feature engineering not required

## 12. Disadvantages

* Requires large data
* Computationally expensive
* Hard to interpret (black box)
