# Assignment

## Step 1

- Open the Anaconda Prompt
Windows: Click the Start menu, type Anaconda Prompt, and click the application. (It looks like a black terminal window with a green icon).

- macOS/Linux: Open your standard Terminal. (If Anaconda is installed, your prompt will usually start with (base)).

## Step 2

 Activate Your Environment (Optional)
If you have created a specific virtual environment for your project, activate it first. If not, you can skip to step 3 to use the default "base" environment.

- Type: conda activate your_env_name

- Press Enter.

## Step 3

Navigate to Your Project Folder
It is a good habit to move to the folder where you want to save your work before launching Jupyter.

- Use the cd (change directory) command. For example:

- cd Documents\DataProjects

- Press Enter.

## Step 4

Launch Jupyter Notebook
Type the following command and press Enter:

- jupyter notebook

## Step 5

Access the Interface
- The Server Starts: You will see lines of text appear in the prompt indicating the server is running.

- Browser Opens: Your default web browser will automatically open a new tab showing the Jupyter dashboard.

- Manual Access: If the browser doesn't open automatically, look at the text in the prompt. Copy the URL starting with http://localhost:8888/... and paste it into your browser.
