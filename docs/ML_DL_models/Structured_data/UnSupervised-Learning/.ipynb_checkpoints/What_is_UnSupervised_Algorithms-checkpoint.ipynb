## UnSupervised Algorithms

### What is an Unsupervised machine learning?
`Unsupervised machine learning` is the process of extracting **patterns** from data. It deals with unlabeled data sets. This form of ML allows algorithms to work on their own to discover information from data, without any guidance. The primary tasks of the Algols here is to classify unsorted informagtion according to patterns, differences, or similarities. `Example` a categorization could be done based on `age, average time spent in hospital, types of disease, etc`.
The final categorization of data set helps to derive some conclusions based on different patterns generated from the data.

### Why Unsupervised machine learning?
Learning from the data is the goal of `Unsupervised machine learning`. The machine derives insight from the data without being thought anything.
- Extracts many kinds of previously unknown patterns in a data set.
- Helps find features that may be useful for categorization.

### Types of Unsupervised Algorithms?
Unsupervised Algorithms extracts patterns from unlabeled or uncategorized data. They used clustering and association techniques help to find the structure of a dataset and group them based on similarity.

### CLUSTERING?
Clustering aims to divides a set of data points into multiple clusters such that the data point withing or cluster are similar to each other. In essence, clustering aims to segregate data points with similar traits.

### ASSOCIATION?
An association rule involves finding the relationships between variables in a large data set. It determines the dependencies of one data item over another data item. Association rule learning often finds it's application in market basket analysis.

### Commonly Used Algorithms?
**Unsupervised Learning Algorithms are:**
- K-Means clustering
- K-memoids
- Desity-based spatial clustering of application with noise (DBSCAN)
- Principal Component Analysis (PCA)
- Agglomerative clustering
- BIRCH

## K-Means clustering
* This is an iterative clustering algorithm that groups the dataset into `k` pre-defined non-pverlapping clusters.
---
## K-medoids
* Is similar to k-means but with the requirments that `cluster centers coincide with points in the data`
---
## DBSCAN
* This algorithms finds core samples of high density and expands clusters from them. It is suitable for a dataset containing clusters of similar density.
---
## PCA
* This algorithms identifies patterns in a dataset and simplifies the dataset by reducing dimensionality.
---
## Agglomerative clustering
* This algorithms forms a hierarchy of clusters by treating each data point as a seperate cluster, this is repeated iteratively untill all the clusters have been merged into one big cluster.
---
## BIRCH
* Is a memory efficient online-learning algorithm. It constructs a tree data structure with cluster centroids being read off the leaf.

### When to use Unsupervised Algorithms?
`Unsupervised Algorithms` are useful when the goal is to extract data from unlabelled data and the output is `NOT` known. They are mainly used when:
* Data points must be grouped based on similar traits.
* The relationship between different features of large dataset should be determined
* The Data input in a large dataset with many features must be reduced to a manageable size i.e the dimensionality of the input data must be reduced.
* Anomalies in dataset should be detected.
---
### Applications of Unsupervised Learning Algorithms?
Unsupervised Learning Algorithms has many real-world applications including:
* *Market basket analysis*
* *Fraud detection*
* *Malware detection*
* *Customer Segmentation*
* *Targeted marketting campaigns*

Thus, the ability of `Unsupervised Algorithms` to group unsorted data based on patterns, similarities and differences make them an ideal choice for application.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\M.faisal\Desktop\ai_ml\ai_ml_projects\datasets\pokemon.csv")
df.head()
```

```python
#Let's drop the categorical varibles as we may not be able to plot them
types = df['Type 1'].isin(['Grass', 'Fire', 'Water'])
drop_cols = ['Type 1', 'Type 2', 'Generation', 'Legendary']
df = df[types].drop(columns=drop_cols)
df.head()
```

### Environment setup
**What this does**:
- Sets a system-level variable that controls how many CPU threads are used by low-level libraries (MKL/OpenMP). Forces computation to run on 1 thread.

**Why it matters**:
- Prevents the KMeans memory leak warning on Windows.
- Slightly reduces parallel performance, but avoids instability.

### Import libraries
**What this does**:
- Imports the KMeans clustering algorithm.
- Imports NumPy for numerical operations (though not heavily used here).

**Parameters**:
- n_clusters=3 → You are asking the model to group data into 3 clusters.
- random_state=3 → Fixes randomness for reproducibility.
  
**Under the hood**: KMeans will:
- Randomly initialize 3 centroids

**Iteratively**:
- Assign points to nearest centroid
- Recompute centroids
- Stop when convergence is reached

```python
import warnings
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")

from sklearn.cluster import KMeans
import numpy as np

#Initialize the KMeans Algorithms
kmeans = KMeans(n_clusters=3, random_state=3)

# Train the model
# Feeds 2 features into the model: ``[Attack, Defense]``
# Internally: Each row becomes a point in 2D space: `(Attack, Defense)`
kmeans.fit(df[['Attack', 'Defense']])

# Assign cluster labels
# Adds a new column cluster to your dataframe. Each row gets a label: `0, 1, or 2`
df['cluster'] = kmeans.labels_

#Extract centroids: It contains a What this contains a: (3 × 2) array: where Each row = one cluster center.
centroids = kmeans.cluster_centers_

# Split centroid coordinates
# Separates centroid coordinates into: `cen_x → Attack values of centroids` and `cen_y → Defense values of centroids`
cen_x = [i[0] for i in centroids]
cen_y = [i[1] for i in centroids]

#Map centroids to each data point
# For each row: Looks at its cluster label, Assigns the corresponding centroid coordinates
# It Helps with visualization (e.g., drawing lines to centroids), and also lets you compare point vs cluster center
df['cen_x'] = [centroids[i][0] for i in df['cluster']]
df['cen_y'] = [centroids[i][1] for i in df['cluster']]
# df['cen_x'] = df.cluster.map({0: cen_x[0], 1:cen_x[1], 2:cen_x[2]})
# df['cen_y'] = df.cluster.map({0: cen_y[0], 1:cen_y[1], 2:cen_y[2]})

#Assign colors to clusters
colors = ['#DF2020', '#81DF20', '#2095DF']
# Map colors to each row
df['c'] = df.cluster.map({0: colors[0], 1:colors[1], 2:colors[2]}) #Assigns a color per cluster
```

```python
# Import plotting library
import matplotlib.pyplot as plt

# A Jupyter magic command (not standard Python). Forces plots to render inside the notebook output cell.
%matplotlib inline

# Scatter plot
plt.scatter(df.Attack, df.Defense, c=df.c, alpha=0.6, s=10)
```

As you can see above, there are three different clusters created using K-means clusters. 
### Breakdown of arguments:
`df.Attack`:
- X-axis values
- Represents feature 1

`df.Defense:`
- Y-axis values
- Represents feature 2

`c=df.c:`
- Color of each point
- Comes from:
  `df['c'] = df.cluster.map({...})`

```python
# Lets a scatter plot by highlighting datapoint for each cluster
plt.scatter(df.Attack, df.Defense, c=df.c, s=df.Speed, alpha=0.6)
```

```python
#Let's create a centroid point for each of the clusters
# Import legend utility
from matplotlib.lines import Line2D

# Create figure and axes
# Creates a figure canvas and an axes object (ax). figsize=(8,8) → square plot.
fig, ax = plt.subplots(1, figsize = (8,8))

# Plot data points
# Plots all data points: `X → Attack, Y → Defense, Color → cluster color (df.c)`
ax.scatter(df.Attack, df.Defense, c=df.c, s=10, alpha=0.6)

# Plot centroids
# Plots cluster centers: `(marker='^' → triangle marker, c=colors → same color as clusters, s=70 → larger size for visibility`
ax.scatter(cen_x, cen_y, marker='^', c=colors, s=70)

# Draw lines from points → centroids
# Iterates over every row in the dataset: For each point; It creates a line from: (`(Attack, Defense) → (cen_x, cen_y)`)
for idx, val in df.iterrows():
    x= [val.Attack, val.cen_x]
    y= [val.Defense, val.cen_y]
    ax.plot(x, y, c=val.c, alpha=0.6)

#Create centroid legend entries; NOTE:(Created but never used)
# cent_leg = [Line2D([0], [0], marker='^', color='w', label='Centroid -c{}'.format(i+1), 
#                   markerfacecolor=mcolor, markersize=10) for i, mcolor in enumerate(colors)]

#Create cluster legend entries
legend_elements = [Line2D([0], [0], marker='^', color='w', label='Cluster {}'.format(i+1), 
                  markerfacecolor=mcolor, markersize=10) for i, mcolor in enumerate(colors)]

# Extend legend with centroids
legend_elements.extend( [Line2D([0], [0], marker='^', color='w', label='Centroid -c{}'.format(i+1), 
                  markerfacecolor=mcolor, markersize=10) for i, mcolor in enumerate(colors)])

# Display legend
ax.legend(handles=legend_elements, loc='upper right', ncol=2)
# Set axis limits, Fixes plot range, Keeps visualization consistent, Prevents auto-scaling
ax.set_xlim(0, 200)
ax.set_ylim(0, 200)
# Title and labels
ax.set_title('Pokemon Cluster', loc='left', fontsize=22)
ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
```

From the above visualization, we can see how data is spread away from the `centroid`. You can use the convex method instead of creating a centroid.
Instead of visualizing clusters using centroids + lines, you can represent each cluster by its outer boundary shape. That boundary is called a `convex hull.`
### Definition (intuitive)

A convex hull is the smallest polygon that encloses all points in a cluster. Think of it like:
- Wrapping a rubber band around all points → that shape is the convex hull

### Difference from centroids:
**The current method (centroid-based):**
**Shows**:
- Cluster center
- Distance from each point to center

**Focus**: distance minimization

**Convex hull method**:
**Shows**:
- The shape and spread of each cluster

**Focus:**
- cluster boundary / geometry

## Performance Parameters
Machine Learning is the science of creating predictive models based on data. Various metrics are used to evaluate the performance of a model. The parameters or performance metrics is chosen based on the type of problem, such as:

- Supervised learning problem
- Unsupervised learning problem
We know that the end goal of `Unsupervised learning algorithms` is to create clusters of similar objects. A model is deemed to perform well, if the members of a given cluster exhibit similar traits.
### Metrics used to evaluate the performance of clustering algorithms are:
- `Silhouette Coefficient`
- `Dunn's index`

1. `Silhouette Coefficient`: The `Silhouette Coefficient` is a measure of similarity of a sample to it's own cluster when compared to other clusters. The `value` of the coefficients ranges from `-1 to +1`. A lower value indicates that the sample is less cohesive(related/connected) to it's own cluster, A higher value indicates a high cohesion. The `Coefficient` is the average distance between a sample and all the remaining points in the same cluster, divided by the average distance between the sample and all the points in the next closest cluster.
 `S = b-a/max(a, b)` where; `a` is the average distance between the sample and all the remaining points in the same cluster; `b` is the average distance between the sample and all the points in the next closest cluster.
`Silhouette Coefficient` for a group of samples is the average of the `Silhouette Coefficient` for each sample. The score lies between `+1` and `-1`.
A score of +1 indicates a highly dense cluster and -1 indicates an incorrect cluster. It the `score` is around `0` then we can infer that the clusters overlap. The higher the score, the denser and well-seperated the cluster will be.

2. `Dunn's index` (`DI`): Is used to determine sets of clusters that are cohesive and well-seperated. A higher `DI` value indicates better clustering. `Dunn's index` can be obtained by dividing the minimum inter-cluster distance by the maximun intra-cluster distance.

Both the `Silhouette Coefficient and Dunn's index`are internal performance evaluation schemes where the results is based on the clustered data itself.

## Hierarchical clustering Analysis (HCA)

Hierarchical clustering Analysis (HCA) is an Unsupervised clustering algorithm. It adopts either a bottom to top approach to grouping data known as `Agglomerative HCA` or a top to bottom approach known as `Divisive HCA`. At first every data in the data point is deemed to be a cluster first. 
As similarities are observed in the closest pairs of data points, objects are added to a specific cluster i.e Objects deemed similar are grouped into clusters. The `endpoint` of the algorthm is reached when every cluster is distict from every other cluster.
`Example`: Say you have a data set of `n`  different type of animals:

- It starts by assuming that each animal is a distinct cluster by itself, that is, n clusters.
- Take some two closest data point and make them a cluster e.g cats and dogs.
- Repeats the process. Cluster mammals, reptiles, fish and so on into different clusters.
- Then mammals, reptiles, fish can then be grouped into the vertebrae cluster. Similarly, an invertebrae cluster can be formed of insects, Corals and Arachnids.
- Finally, there's one large cluster of animals as shown in the diagram below(`Dendrogram`);
