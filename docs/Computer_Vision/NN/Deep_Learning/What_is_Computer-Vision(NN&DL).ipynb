## What is Neural Network?

A **neural network** is a computational model inspired by the structure and function of the **human brain**. It consists of interconnected layers of artificial "neurons" (also called nodes) that process information and learn to recognize complex patterns from data.

### 1. Structure the layers

Neural networks are typically organized into three main types of layers: 

- **Input Layer**: The first layer that receives raw data (like image pixels or text data) and introduces it to the network.
- **Hidden Layers**: One or more intermediate layers where the actual "learning" happens. These layers extract features and perform complex mathematical transformations.
- **Output Layer**: The final layer that delivers the system's prediction or classification (e.g., "this image is a car").

Each neuron performs:
[
z = wx + b
]
[
a = f(z)
]

### 2. Apply weights and biases 
Each connection between neurons has a **weight**, which determines the influence of one neuron on another. A **bias** is an additional value added to the calculation, acting as a **threshold** that helps the model shift its decisions to better fit the data.

### 3. Use activation functions
After a neuron calculates its weighted sum, it passes the result through an activation function (like ReLU or Sigmoid). This function determines whether a neuron should "fire" or send its signal forward, allowing the network to handle nonlinear relationships in data.

### 4. Train via backpropagation 
The network improves through a process called training, which relies on two main steps: 
- **Forward Pass**: The data travels from input to output to generate a prediction.
- **Backpropagation**: The network compares its prediction against the correct answer (calculating the loss), then travels backward to adjust its **weights** and **biases** to reduce future errors.

## Common Types of Neural Networks
- **Feedforward Neural Networks (FNN)**: The simplest type where data moves in only one direction, from input to output.
- **Convolutional Neural Networks (CNN)**: Specialized for images and grid-like data.
- **Recurrent Neural Networks (RNN)**: Designed for sequential data (like speech or time series) by using internal memory loops.
- **Transformers**: A modern architecture that processes whole sequences at once, powering large language models like ChatGPT.

## What is Deep Learning?

**Deep Learning** is a subset of Machine Learning that uses **Artificial Neural Networks (ANNs)** with multiple layers to model complex patterns in data.

It is especially powerful for:

* Image recognition
* Natural Language Processing (NLP)
* Speech recognition
* Time-series forecasting

### Why Deep Learning?

Traditional ML struggles with:

* High-dimensional data
* Unstructured data (images, text, audio)

Deep Learning excels because it:

* Automatically extracts features
* Scales with more data
* Learns hierarchical representations

## Deep learning frameworks

Deep learning frameworks are specialized software libraries that provide the essential building blocks—such as pre-defined neural network layers, optimizers, and hardware acceleration—needed to design, train, and deploy complex AI models without writing everything from scratch. 

### Top Deep Learning Frameworks

* **PyTorch**: Currently the leading choice for academic research and rapid prototyping due to its dynamic computation graph and "Pythonic" design. It was developed by Meta's AI Research lab and is highly flexible.
  
* **TensorFlow**: Developed by Google Brain, this is a mature, production-grade framework often preferred for large-scale industrial deployments. It offers robust tools like TensorFlow Extended (TFX) for production pipelines and TensorFlow Lite for mobile/edge devices.

* **Keras**: A user-friendly, high-level API that simplifies building neural networks. While it is now integrated directly into TensorFlow (as tf.keras), it also serves as a multi-backend interface that can run on top of PyTorch or JAX.

* **JAX**: A high-performance framework from Google designed for high-performance scientific computing and machine learning. It is increasingly popular for large-scale model training due to its advanced automatic differentiation and TPU support.

* **MXNet**: Supported by the Apache Software Foundation and preferred by Amazon for AWS, it is known for high scalability in distributed computing environments.

### Which One Should You Choose?
The best framework often depends on your specific goals and technical background:

| If you are... [2, 4, 5, 6, 7, 11, 14, 15] | Recommended Framework | Key Reason |
|---|---|---|
| A Beginner | Keras[](https://keras.io/) | Extremely simple API and low-code interface. |
| A Researcher | PyTorch[](https://pytorch.org/) | Flexible, dynamic graphs make experimentation and debugging easier. |
| In Industry | TensorFlow[](https://www.tensorflow.org/) | Superior deployment tools and scalability for large commercial systems. |
| Handling Big Data | BigDL | Deep learning specifically optimized for Apache Spark and Hadoop clusters. |
