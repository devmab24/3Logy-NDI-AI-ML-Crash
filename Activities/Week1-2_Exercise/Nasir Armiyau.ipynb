# Running Jupyter Notebook from the env Folder (on Desktop) By Nasir Armiyau
## Objective

> This guide explains the step-by-step process for launching Jupyter Notebook from a virtual environment located on the Desktop.
> 
## Step-by-Step Procedure
### 1️⃣ **Open Anaconda Prompt**

- Step 1: Press the Windows key on your keyboard.

- Step 2: Type Anaconda Prompt.

- Step 3: Right-click on Anaconda Prompt → select Run as Administrator.

- Step 4: Wait for the terminal window to open.

  ### 2️⃣ Navigate to the Desktop

- Step 5: Type the following command:

  **cd "C:\Users\Maihanci\OneDrive\Desktop\"**

- Step 6: Press Enter.

  ### 3️⃣ Move into the Project Folder

- Step 7: Enter the project folder by typing:

**cd ai_ml_project**

- Step 8: Press Enter.

  ### 4️⃣ Activate the Virtual Environment

- Step 9: Activate the virtual environment by typing:

**conda activate "C:\Users\Maihanci\OneDrive\Desktop\ai_ml_project\env"**

- Step 10: Press Enter

- Step 11: Confirm activation — you should see (env) at the beginning of the command line.

  ### 5️⃣ Launch Jupyter Notebook

- Step 12: Start Jupyter Notebook by typing:

**jupyter notebook**

- Step 13: Press Enter

- Step 14: Your default browser should open automatically with Jupyter Notebook.

> 💡 Note: If it doesn’t open automatically, copy the link from the Command Prompt and paste it into your preferred browser.
> ### 6️⃣ Create a New Notebook

- Step 15: Click New

- Step 16: Select Python 3 to create a new notebook.

  ### ✅ After Finishing

- Step 17: Close the Jupyter browser tab.

- Step 18: Return to Anaconda Prompt

- Step 19: Type:

**deactivate**

- Step 20: Press Enter to exit the virtual environment.

  > Bye Bye!
