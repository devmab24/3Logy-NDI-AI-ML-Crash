## The two main data structures(datatypes) in the pandas library are 
- **Series** and **DataFrame**.

1. **Pandas Series**: A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, etc.). 

- **Dimensions**: 1D.
- **Analogy**: Similar to a single column in an Excel spreadsheet or a Python list with labels.
- **Key Feature**: Every element has an associated label called an index.

Read more: https://pandas.pydata.org/docs/reference/api/pandas.Series.html

## Tabular data
- **Series**

````python
# Creating a Series (1-dimensional i.e 1 column)
# prices = pd.Series([100, 250, 450, 10, 25, 45, 130, 220])
# prices

# Creating a Series (1-dimensional i.e 1 column)
fruits_types = pd.Series(["Oranges", "lemons", "limes", "Watermelon", "honeydew", "cantaloupe", "Strawberries", "blueberries", "raspberries"], name="Fruits")
fruits_types
````

````python
# Creating a Series (1-dimensional i.e 1 column)
fruit_colour = pd.Series(["Orange", "Green", "Blue", "Yellow", "Black", "Gray", "Red", "Pink"], name="Colours")
fruit_colour
````

````python
# Using pd.concat() along the column axis (axis=1)
df_concat = pd.concat([fruits_types, fruit_colour], axis=1)
print("\nDataFrame using pd.concat(axis=1):")
print(df_concat)
````

````python
# # Create multiple Series
# s1 = pd.Series([10, 20, 30, 40], name="Numbers")
# s2 = pd.Series(["apple", "orange", "banana", "grape"], name="Fruits")
# s1
# s2

# # Method 1: Using a dictionary of Series (Recommended)
# data_dict = {"Numbers": s1, "Fruits": s2}

# df_dict = pd.DataFrame(data_dict)
# print("\nDataFrame from a dictionary of Series:")
# print(df_dict)

# Method 2: Using pd.concat() along the column axis (axis=1)
# df_concat = pd.concat([s1, s2], axis=1)
# print("\nDataFrame using pd.concat(axis=1):")
# print(df_concat)

````

**NB**: The axis parameter defines the direction of the "glue." It is the most important concept to master in pandas. 
- **axis=0** (Default): (Vertical), Stacks the second object under the first (adds rows).
- **axis=1**: (Horizontal), Stacks the second object to the right of the first (adds columns).

- **DataFrame**

2. **Pandas DataFrame**: A DataFrame is a two-dimensional labeled data structure with rows and columns. It's far more common.

- **Dimensions**: 2D.
- **Analogy**: Similar to an entire Excel sheet or a SQL table.
- **Key Feature**: It can be thought of as a dictionary-like container for Series objects, where each column is its own Series.

````python
# Creating a DataFrame
Fruits = pd.DataFrame({"Fruit Name": fruits_types, "Fruit Color": fruit_colour})
Fruits
````

## Manipulating Data

````python
 car_sales_df.head()
````

````python
car_sales_df["brand"].str.lower()
````

````python
# If you try to access the str on numerical value you will get an error `AttributeError`
# car_sales_df["year"].str.lower()
````

````python
# If you print the brand column you'll notice that the lower case isn't applied thats because you have'nt re-assign it
car_sales_df["brand"]

# After re-assignment
car_sales_df["brand"] = car_sales_df["brand"].str.lower()
car_sales_df["brand"]
````

````python
# Some function have parameters to make changes and effect the change without reassigning; let's see examples with missing data
#NB: real datasets often contains missing data's
car_sales_missing_df = pd.read_csv("datasets/car_datasets/car_sales_missing.csv")
car_sales_missing_df.info()
````

````python
#You'll see that some rows have a `NaN`; which means not a number; thats a way python tells you that such cell has a missing value
car_sales_missing_df
````

````python
# fillna
car_sales_missing_df["year"].fillna(car_sales_missing_df["year"].mean())
````

````python
car_sales_missing_df["year"].info()
````

````python
car_sales_missing_df["quantity_sold"].fillna(car_sales_missing_df["quantity_sold"].mean(), inplace=True)
(car_sales_missing_df["quantity_sold"]).info()
````
