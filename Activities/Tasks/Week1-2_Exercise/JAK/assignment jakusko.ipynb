## **Project Setup: AI/ML Development Environment**

This guide covers the standard procedure for initializing an isolated **Conda** environment and launching **Jupyter Notebook**.

### **Step 1: Open the Anaconda Prompt**

The Anaconda Prompt is a CLI pre-configured with the Conda package manager.

1. Press the **Windows Key**.
2. Type **"Anaconda Prompt"** and press Enter.
3. **Verify:** Look for the `(base)` prefix in the terminal window.

---

### **Step 2: Navigate to Project Folder**

You must move the terminal focus to your specific project directory to ensure files are saved in the correct location.

**Run the following command:**

```bash
cd C:\Users\adamu\Desktop\AI_ML001

```

---

### **Step 3: Activate the Local Environment**

Activating the environment ensures you are using the specific Python version and library set required for this project.

**Run the activation command:**

```bash
conda activate C:\Users\adamu\Desktop\AI_ML001\env

```

> **Note:** Your command prompt prefix should now change from `(base)` to `(env)` (or the specific path to your environment).

---

### **Step 4: Launch Jupyter Notebook**

This initiates a local web server (usually on port `8888`) and opens the interface in your default browser.

**Run the command:**

```bash
jupyter notebook

```

---

### **Step 5: Create a New Notebook**

Once the browser opens to `http://localhost:8888`:

1. Click the **New** button (top-right).
2. Select **Python 3 (ipykernel)**.
3. A new `.ipynb` file will open in a new tab.

---

### **Step 6: Proper Shutdown (Critical)**

Closing the browser tab **does not** stop the Python kernel or the server.

| Task | Action |
| --- | --- |
| **Step A** | Save and Close the notebook tab in your browser. |
| **Step B** | Return to the **Anaconda Prompt** window. |
| **Step C** | Press **`CTRL + C`** on your keyboard. |
| **Step D** | Type **`Y`** when asked to terminate, then press **Enter**. |
