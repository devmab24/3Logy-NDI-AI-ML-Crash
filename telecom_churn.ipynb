```python
custumer_churn['TotalCharges'] = pd.to_numeric(custumer_churn['TotalCharges'], errors='coerce')
custumer_churn.head()
```

```python
custumer_churn['TotalCharges'] = custumer_churn['TotalCharges'].fillna(custumer_churn['TotalCharges'].median())
custumer_churn.drop('customerID', axis=1, inplace=True)
```

```python
custumer_churn['TotalCharges'].fillna(custumer_churn['MonthlyCharges'], inplace=True)
```

```python
custumer_churn['Churn'] =custumer_churn['Churn'].map({'Yes': 1, 'No': 0})
```

```python
cols_to_fix = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup',
               'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in cols_to_fix:
    custumer_churn[col] = custumer_churn[col].replace({'No internet service': 'No', 'No phone service': 'No'})
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
import joblib
```

```python
# Separate Numeric and Categorical Columns
num_cols = custumer_churn.select_dtypes(include=['int64', 'float64']).columns
cat_cols = custumer_churn.select_dtypes(include=['object', 'string']).columns

# Preprocessing
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', drop="first"), cat_cols)
])

# Transform
X_processed_scaled = preprocessor.fit_transform(custumer_churn)

# Get correct feature names
feature_names = preprocessor.get_feature_names_out()


custumer_churn.head()
```
