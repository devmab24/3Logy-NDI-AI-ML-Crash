## Open the Anaconda Prompt

### Procedure


- Click the Start Menu in Windows.

- Search for Anaconda Prompt.

- Click to open Anaconda Prompt

### Anaconda prompt

The Anaconda Prompt is a terminal environment configured with the Conda package manager. It automatically activates the base environment, allowing Python and Jupyter commands to run correctly.
    

### Code appearance in Anaconda Prompt

(base) C:\Windows>\Users\murja\Desktop\AI_ML001

### Activate a Conda Environment

(base) C:\Users\murja\Desktop\AI_ML001>conda activate

### Conda Environment

A Conda environment isolates packages and dependencies so that different projects do not conflict with each other. Activating the environment ensures that Jupyter Notebook runs with the correct Python version and libraries.

After activation, the prompt changed to:

(base) C:\Users\murja\Desktop\AI_ML001>cd \Users\murja\Desktop\AI_ML001

### Navigate to the Desired Working Directory

Move to the folder where you want to store your notebooks.

(base) C:\Users\murja\Desktop\AI_ML001>conda activate C:\Users\murja\Desktop\AI_ML001\env

### cd (Change directory)

The cd (change directory) command sets the working directory. When Jupyter Notebook launches, it will display and save files in this location.

### Launch Jupyter Notebook

(C:\Users\murja\Desktop\AI_ML001\env) C:\Users\murja\Desktop\AI_ML001>jupyter notebook

### When the command is executed

- The system starts the Jupyter Notebook server.

- A local web server is created on your computer.

- The server usually runs on:

http://localhost:8888

## Create a New Notebook

- Click New

- Select Python 3 (ipykernel)

### Stop the Jupyter Notebook Server

When you finish working, the server should be shut down properly.

- Return to the Anaconda Prompt window.

- Press:

CTRL + C
