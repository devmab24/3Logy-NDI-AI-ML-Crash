---

### 21. Differentiate between Machine Learning and Deep Learning.

<details>
<summary>Show Answer</summary>

Machine Learning often relies on manual feature engineering, while Deep Learning automatically learns features using neural networks.

</details>

---

### 22. What is overfitting in Deep Learning?

<details>
<summary>Show Answer</summary>

Overfitting occurs when a model memorizes training data and performs poorly on unseen data.

</details>

---

### 23. What is underfitting?

<details>
<summary>Show Answer</summary>

Underfitting occurs when a model cannot properly learn patterns from data.

</details>

---

### 24. Explain the purpose of hidden layers.

Hidden layers help neural networks learn complex patterns and representations.

</details>

---

### 25. What is forward propagation?

<details>

</details>

---

### 26. What is backpropagation?

<details>

</details>

### 27. What is a loss function?

</details>

### 28. What is an optimizer?
