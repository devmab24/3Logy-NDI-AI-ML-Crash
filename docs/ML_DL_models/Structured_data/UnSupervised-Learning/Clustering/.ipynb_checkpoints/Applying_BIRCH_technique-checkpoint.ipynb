# Applying BIRCH (Balance Iterative Reducing and Clustering)
An Unsupervised technique

```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
```

```python
X, clusters = make_blobs(n_samples=450, centers=6, cluster_std=0.7, random_state=0)
plt.scatter(X[:, 0], X[:, 1], alpha=0.7, edgecolors='b')
```

```python
#Now, let's implement BIRCH
brc = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
brc.fit(X)
```

```python
labels = brc.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```

As you can see, the algorithms has created 6 different clusters based on the datasets. This is how you can implement Birch for creating different clusters.
