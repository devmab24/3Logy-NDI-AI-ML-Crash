# E. Sequence
**Sequence** (`list`, `tuple`, `set`)**:**

- Data that is a collection of discrete items.
- A `list` is collection that is ordered and changeable. It's designated using square brackets `[]`, and items can be of different data types: `["red", 1, 1.03, 1]`.
- A `tuple` is a collection which is ordered and unchangeable. It's designated using parentheses `()`: `("red", 1, 1.03, 1)`.
- A `set` is a collection which is unordered, unchangeable, and does not permit duplicate items. It's designated using curly brackets `{}`: `{"red", 1, 1.03}`.

## Key Characteristics
1. List [ ]
- **Mutable:** You can add, remove, or modify elements after creation using methods like append(), remove(), or direct index assignment (e.g., my_list[0] = 5).
- **Ordered:** Elements maintain their insertion order, and you can access them using integer indices starting from 0.
- **Best Use Case:** When you have a collection of items that may need to grow, shrink, or be sorted during your program's execution.

2. Tuple ( )
- **Immutable:** Once created, you cannot change its content. This makes them safer for data that should remain constant.
- **Performance:** Slightly faster and more memory-efficient than lists because their size is fixed.
- **Hashable:** Because they are immutable, tuples can be used as keys in dictionaries, whereas lists cannot.
- **Best Use Case:** For fixed data like geographical coordinates, RGB color values, or constant configuration settings.

3. Set { }
- **Unique Elements:** Sets automatically remove any duplicate values. If you try to add a duplicate, it is simply ignored.
- **Unordered:** There is no defined order for elements; therefore, you cannot access items by index (e.g., my_set[0] will cause an error).
- **Mathematical Operations:** Sets support powerful operations like unions, intersections, and differences.
- **Best Use Case:** For removing duplicates from a collection or performing high-speed membership tests (checking if x in collection).

````python
my_list  = [1, 2, 2, 3]  # Output: [1, 2, 2, 3]
my_tuple = (1, 2, 2, 3)  # Output: (1, 2, 2, 3)
my_set   = {1, 2, 2, 3}  # Output: {1, 2, 3} (Duplicates removed)
````

## 1.1 List Comprehension
**Definition:** List comprehension in Python is a concise and efficient way to create a new list by applying an expression or a filter to items from an existing iterable (like a list, tuple, range, or string). It offers a more readable and compact syntax compared to traditional for loops and is often faster.

**Why it matters in ML:** Used for normalization and feature scaling.


## Syntax
The basic syntax of a list comprehension is a single line of code enclosed in square brackets []:


new_list = [expression for item in iterable if condition]

- **expression:** The output value or operation applied to each item.
- **for item in iterable:** Defines the iteration source.
- **if condition (optional):** Filters items, including only those where the condition is True.

````python
# Creates a list of squares: [0, 1, 4, 9, ..., 81]
squares = [x**2 for x in range(10)]
````

````python
numbers = [1, 2, 3, 4]
squared = [x**2 for x in numbers]
print(squared)
````

# G. Loops (Control flow statements)
**Definition:** loops are control flow statements that execute a block of code repeatedly. The two main types are for loops and while loops, each suited for different scenarios.

**Why it matters in ML:** Used for training epochs and iterating over data.


**For Loops**: 
A for loop in Python iterates over the items of any sequence (such as a list, tuple, string, or range) or other iterable objects, executing the loop body once for each item.

**How it works:**
- **Iteration Variable**: A temporary variable (e.g., item in the syntax below) is assigned the value of the current item in the sequence during each iteration.
- **Sequence Access**: Python internally uses an iterator to access elements one by one in the order they appear.
- **Termination**: The loop continues until all items in the sequence are processed. Once the sequence is exhausted, the loop terminates.

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I love {fruit}!")
````

````python
for epoch in range(5):
    print('Epoch:', epoch)
````
