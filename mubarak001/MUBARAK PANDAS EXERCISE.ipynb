Now let's deal with missing data.

```python
# Import the car sales DataFrame with missing data ("car-sales_missing.csv")

# Check out the new DataFrame
import pandas as pd

# 1. Define the path (using 'r' before the string helps with Windows backslashes
file_path = r"C:/Users/msheh/Desktop/AI_ML001/3Logy-NDI-AI-ML-Crash/datasets/car_sales_missing.csv"

# 2. Use pd.read_csv to actually load the data
car_sales_missing = pd.read_csv(file_path)

# 3. View the result
car_sales_missing
```

Notice the missing values are represented as `NaN` in pandas DataFrames.

Let's try fill them.

```python
# Fill the quantity column missing values with the mean of the column inplace
# The modern way: Re-assign the column to save the changes
car_sales_missing["quantity_sold"] = car_sales_missing["quantity_sold"].fillna(car_sales_missing["quantity_sold"].mean())
```

```python
# View the car sales missing DataFrame and verify the changes
# View the first 10 rows to see the filled values
car_sales_missing.head(10)
```

```python
# Remove the rest of the missing data inplace
# This does the exact same thing but is more "future-proof"
car_sales_missing = car_sales_missing.dropna()
```

```python
# Verify the missing values are removed by viewing the DataFrame
car_sales_missing.head()
```

We'll now start to add columns to our DataFrame.

```python
# Create a "Seats" column where every row has a value of 5
car_sales_missing["Seats"] = 5
car_sales_missing.head()
```

```python
# Create a column called "Engine Size" with random values between 1.3 and 4.5
# Remember: If you're doing it from a Python list, the list has to be the same length
# as the DataFrame
import numpy as np

# 1. Generate random floats between 1.3 and 4.5
# We use len(car_sales) to ensure the list matches the DataFrame length exactly
car_sales_missing["Engine Size"] = np.random.uniform(1.3, 4.5, size=len(car_sales_missing))
car_sales_missing.head()
```

```python
# Create a column which represents the price of a car per kilometer
# Then view the DataFrame
# Rename 'quantity_sold' to 'odometer' permanently
car_sales_missing.rename(columns={"quantity_sold": "odometer"}, inplace=True)

# Confirm the name change
print(car_sales_missing.columns)
car_sales_missing
```

```python
# Remove the last column you added using .drop()
car_sales_missing.drop("odometer", axis=1, inplace=True)
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the the shuffled DataFrame to a new variable
car_sales_missing_shuffled = car_sales_missing.sample(frac=1)
car_sales_missing_shuffled
```

Notice how the index numbers get moved around. The [`sample()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html) function is a great way to get random samples from your DataFrame. It's also another great way to shuffle the rows by setting `frac=1`.

```python
# Reset the indexes of the shuffled DataFrame
car_sales_missing_shuffled.reset_index(drop=True, inplace=True)
car_sales_missing_shuffled
```

Notice the index numbers have been changed to have order (start from 0).
