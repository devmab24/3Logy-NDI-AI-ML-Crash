# TEXT PREPROCESSING

## Text Preprocessing

Text preprocessing is the process of cleaning and preparing text data for analysis.

Steps include:
- Tokenization
- Lowercasing
- Removing stopwords
- Stemming / Lemmatization

## Tokenization with NLTK

In the Natural Language Toolkit (NLTK), *tokenization* is the process of breaking down raw text into smaller, meaningful units called tokens. It is typically the first step in a Natural Language Processing (NLP) pipeline, transforming unstructured character strings into a structured list that computers can analyze.

```python
from nltk.tokenize import word_tokenize

text = "Artificial intelligence (AI) is a BRANCH of computer science that creates systems capable of performing tasks that typically require human intelligence, such as learning, reasoning, problem-solving, perception, and decision-making"

tokens = word_tokenize(text)
print(tokens)
```

## Lowercasing

Lowercasing in the context of NLTK (Natural Language Toolkit) is a fundamental text preprocessing step that involves converting all characters in a dataset to lowercase to ensure uniformity.

```python
tokens_lower = [word.lower() for word in tokens]
print(tokens_lower)
```

## Removing Stopwords

In Natural Language Processing (NLP), removing stopwords with NLTK is the process of filtering out common, high-frequency words that typically carry little unique semantic meaning. This is a core text preprocessing step intended to focus computational resources on the "keywords" that define a sentence's actual context.

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered = [word for word in tokens_lower if word not in stop_words]
print(filtered)
```

## Lemmatization with spaCy

Lemmatization in spaCy is the process of reducing inflected or varied forms of a word (e.g., "running", "ran") down to its base dictionary form, known as the lemma (e.g., "run"). Unlike stemming, spaCy uses Part-of-Speech (POS) tagging to ensure the lemma is a valid, meaningful word. It is a default component in the spaCy pipeline, accessible via the .lemma_ attribute.

```python
doc = nlp(text)

lemmatized = [token.lemma_ for token in doc]
print(lemmatized)
```

# SENTIMENT ANALYSIS

## Sentiment Analysis

Sentiment Analysis determines whether a text is:
- Positive
- Negative
- Neutral

Used in:
- Product reviews
- Social media analysis

## NLTK Sentiment

In NLTK, Sentiment Analysis refers to the computational process of determining whether a piece of text expresses a positive, negative, or neutral emotional tone.

```python
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

text = "I love learning AI, it's amazing!"

score = sia.polarity_scores(text)
print(score)
```

```python
# Interpret Result
if score['compound'] > 0:
    print("Positive")
elif score['compound'] < 0:
    print("Negative")
else:
    print("Neutral")
```

# NAMED ENTITY RECOGNITION (NER)

## Named Entity Recognition (NER)

NER identifies entities in text such as:
- Person
- Organization
- Location
- Date

## spaCy NER

In spaCy, NER stands for Named Entity Recognition, which is a powerful feature used to automatically identify and classify "real-world objects" within unstructured text.

```python
text = "Elon Musk is the CEO of Tesla and lives in the USA."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

## Visualization

```python
from spacy import displacy

displacy.render(doc, style="ent", jupyter=True)
```

# INTRO TO GENSIM (BASIC)

Gensim is a free, open-source Python library designed for unsupervised topic modeling and natural language processing (NLP). The name "Gensim" is an abbreviation of "Generate Similar".

## Gensim (Introduction): https://radimrehurek.com/gensim/intro.html

Gensim is widely recognized for being highly optimized, fast, and robust. It is often used by researchers and businesses to extract insights from large volumes of unstructured text data like reviews, articles, or social media posts.

- Topic modeling
- Word similarity
- Document similarity

### Example

```python
from gensim.utils import simple_preprocess

text = "Machine learning is powerful and useful."

tokens = simple_preprocess(text)
print(tokens)
```

# MINI PRACTICE TASK

## Practice Task

1. Take a sentence from the user
2. Perform:
   - Tokenization
   - Remove stopwords
   - Lemmatization
3. Print the cleaned text

```python
user_text = input("Enter a sentence: ")

tokens = word_tokenize(user_text)
tokens = [word.lower() for word in tokens]

filtered = [word for word in tokens if word not in stop_words]

doc = nlp(" ".join(filtered))
lemmatized = [token.lemma_ for token in doc]

print("Processed Text:", lemmatized)
```
